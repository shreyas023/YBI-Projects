 **Dataset 1** - Survival of passengers on Titanic




```python
#import data
import pandas as pd
```

```python
#explore data
titanic = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Titanic.csv')
titanic.head()
```

```python
titanic.info()
titanic.describe()
```

```python
titanic.shape
```

```python
titanic.nunique()
```

```python
titanic.corr()
```

```python
#define y and X
titanic.columns
```

```python
y = titanic['survived']
X = titanic.drop('survived',axis=1)
```

```python
#train test split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=2529,test_size=0.3)
```

**Dataset 2** - Selling price of Headphones on Flipkart

```python
import pandas as pd
```

```python
y = price['Selling Price']
X = price.drop('Selling Price',axis=1)
```
